```python
loan_temp['annual_inc']=loan_temp['annual_inc'].astype(int)
```

```python
##apply binning on annual income [0-30000, 30000-60000, 60000-100000, 100000+]

def bininc(incparam):
    if incparam >=0 and incparam < 30000:
        return '0-30k'
    elif incparam >=30000 and incparam < 60000:
        return '30-60k'
    elif incparam >=60000 and incparam < 100000:
        return '60-100k'
    elif incparam >= 100000:
        return '100k+'

loan_temp['anninc_bin'] = loan_temp.annual_inc.apply(bininc)
```

```python
##findding the correlation between interest rates and default ratio
#start with binning
def binint(intparam):
    
    if intparam >= 0 and intparam < 8:
        return '0-8%'
    elif intparam >= 8 and intparam < 13:
        return '8-13%'
    elif intparam >=13 and intparam < 18:
        return '13-18%'
    elif intparam >=18:
        return '18%+'
loan_temp['int_rate_bin'] = loan_temp.int_rate.apply(binint) 
```

```python
loan_temp.int_rate_bin
```

```python
intdef = loan_temp[loan_temp['defaulted']==1].groupby(['int_rate_bin'])[['defaulted']].count().reset_index()
intdeftot = loan_temp.groupby(['int_rate_bin'])[['defaulted']].count().reset_index()
intdeffull = pd.merge(intdef, intdeftot, left_on='int_rate_bin', right_on='int_rate_bin', how='left')
intdeffull['ratio'] = intdeffull['defaulted_x']/intdeffull['defaulted_y']*100
intdeffull.rename(columns={'defaulted_x':'defaulted', 'defaulted_y':'totals'}, inplace=True)
intdeffull
```

```python
##visualizing the data
ax = sns.barplot(x='int_rate_bin', y='ratio', data=intdeffull, hue='ratio')
```
